```python
query="""
SELECT
    category,
    SUM(films) AS total_films,
    SUM(CASE WHEN rating = 'G' THEN films ELSE 0 END) AS 'G',
    SUM(CASE WHEN rating = 'PG' THEN films ELSE 0 END) AS 'PG',
    SUM(CASE WHEN rating = 'PG-13' THEN films ELSE 0 END) AS 'PG-13',
    SUM(CASE WHEN rating = 'R' THEN films ELSE 0 END) AS 'R'
FROM (
    SELECT
        category.name AS category,
        film.rating,
        COUNT(film.film_id) AS films
    FROM category
    JOIN film_category ON category.category_id = film_category.category_id
    JOIN film ON film_category.film_id = film.film_id
    GROUP BY category.name, film.rating
    ORDER BY films DESC
) sub
GROUP BY category
ORDER BY total_films DESC;
"""


dp=mysql.connector.connect(host='localhost',user='root',password='ziadziad',database='sakila')
cursor=dp.cursor()
cursor.execute(query)
results=cursor.fetchall()
for result in results:
    print(result)
```

```python
#inserting values
query='insert into actor (first_name,last_name) values(%s, %s)'

values=('ahmed','adel')

cursor.execute(query,values)
results=cursor.fetchall()
dp.commit()#to save the value in the tables in dpgate
for result in results:
    print(result)
```

```python
#excute many times
query='insert into actor (first_name,last_name) values(%s, %s)'

values=[('ahmed','adel'),
        ('nada','adel'),
        ('ziad','adel'),
        ('alaa','adel'),
]
cursor.executemany(query,values)

dp.commit()#to save the value in the tables in dpgate

print(cursor.rowcount,'was inserted.')
```

```python
#update 
query='update actor set first_name= %s where first_name=%s '

values=('mohammed','ziad')
cursor.execute(query,values)

dp.commit()#to save the value in the tables in dpgate

print(cursor.rowcount,'was inserted.')
```
