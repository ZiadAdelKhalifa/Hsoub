```python
# Get information from a YouTube channel

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import FirefoxOptions
import time
import tabulate

options = FirefoxOptions()
# options.add_argument('--headless')  # Uncomment this to run in headless mode

def get_youtube_channel_data():
    # Prompt user to input the YouTube channel name
    channel = input('What channel are you looking for: ')  # e.g., mohoobada or HsoubAcademy
    
    with webdriver.Firefox(options=options) as driver:
        # Navigate to the YouTube channel's videos page
        driver.get('https://youtube.com/@' + channel + '/videos')
        time.sleep(3)  # Wait for the page to load

        # Finds the <html> element of the page, which represents the entire webpage
        page = driver.find_element(By.TAG_NAME, 'html')

        # Scroll down the page 3 times to load more videos
        for i in range(3):
            time.sleep(3)  # Wait for more content to load after scrolling
            page.send_keys(Keys.END)
        
        try:
            # Find all video titles and their URLs
            video_titles = driver.find_elements(By.ID, 'video-title-link')

            videos = []
            for title in video_titles:
                videos.append([title.text, title.get_property('href')])  # Store title and URL in a list
                
            # Finds all elements with the ID metadata-line, which contains information like views and publish date
            video_meta_data = []
            videos_met = driver.find_elements(By.ID, 'metadata-line')
            
            for meta_data in videos_met:
                span_data = meta_data.find_elements(By.TAG_NAME, 'span')

                span_text = []

                for span in span_data:
                    span_text.append(span.text)  # Extract text from each span element
                video_meta_data.append(span_text)  # Append metadata text to the list

            # Combine video titles, URLs, and metadata
            results = []
            for item in zip(videos, video_meta_data):
                results.append(item[0] + item[1])  # Concatenate title/URL with metadata

            # Format the results into a table and print
            tabled_data = tabulate.tabulate(results, headers=['Title', 'URL', 'Views', 'Date of Publish'], tablefmt='fancy-grid')
            print(tabled_data)

        except Exception as e:
            print(f'An error occurred: {e}')  # Print any errors that occur

if __name__ == '__main__':
    get_youtube_channel_data()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver import FirefoxOptions
import time
import tabulate

options = FirefoxOptions()
options.add_argument('--headless')

def get_youtube_channel_data():
    channel = input('What channel are you looking for: ')  # mohoobada #HsoubAcademy
    
    with webdriver.Firefox(options=options) as driver:
        driver.get('https://youtube.com/@' + channel + '/videos')
        time.sleep(3)

        page = driver.find_element(By.TAG_NAME, 'html')

        for i in range(3):
            time.sleep(3)
            page.send_keys(Keys.END)
        try:
            content = driver.find_element(By.ID, 'content')
            video_titles = driver.find_elements(By.ID, 'video-title-link')

            videos = []
            for title in video_titles:
                videos.append([title.text, title.get_property('href')])
            
            
            video_meta_data = []
            videos_met = driver.find_elements(By.ID, 'metadata-line')
            
            for meta_data in videos_met:
                span_data = meta_data.find_elements(By.TAG_NAME, 'span')

                span_text = []
                for span in span_data:
                    span_text.append(span.text)
                video_meta_data.append(span_text)
            
            
            results = []
            for item in zip(videos, video_meta_data):
                results.append(item[0] + item[1])  # Fixed the way items are appended
            

            tabled_data = tabulate.tabulate(results, headers=['Title', 'URL', 'Views', 'Date of Publish'], tablefmt='fancy-grid')
            print(tabled_data)

        except Exception as e:
            print('The page was not found:', e)

if __name__ == '__main__':
    get_youtube_channel_data()
```
