```python
import requests
from bs4 import BeautifulSoup
import re
from tabulate import tabulate
from datetime import datetime, date
import json

def get_forcast_data():
    url='https://world-weather.info/'
    #user-agent:to avoid prevent the rquest from the website
    headers={'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
             ,'cookie' :'celsius=1'}

    response=requests.get(url,headers=headers)

    if response.ok:
        soup=BeautifulSoup(response.content , 'html.parser')

        resorts=soup.find('div',id='resorts')
        #print(resorts)

        #\w+ any number of letters \s any number of white spaces
        #get cites names
        re_cities=r'">([\w\s]+)<\/a>'
        cites=re.findall(re_cities,str(resorts))
        print(cites)


        #get the temp values
        re_temp=r'>(\+\d+|-\d+)'
        temps=re.findall(re_temp,str(resorts))
        temps=[int(temp) for temp in temps]
        print(temps)

        #get the weather condition
        condition_tage=resorts.find_all('span',class_='tooltip')
        conditions=[condition.get('title') for condition in condition_tage ]
        print(conditions)

        data=zip(cites,temps,conditions)
        return data
    return False

def get_weather_txt(data):
    if data:
        today = date.today().strftime('%d/%m/%Y')
        with open(file='output.txt', mode='w') as file:
            file.write('Popular Cities Forecast\n')
            file.write(today + '\n')
            file.write('='*30 + '\n')
            table = tabulate(data, headers=['City', 'Temp', 'Condition'], tablefmt='pretty')
            file.write(table)


def get_weather_json(data):
    if data:
        today = date.today().strftime('%d/%m/%Y')
        cities = [{'City': city, 'Temp': temp, 'Condition': condition} for city, temp, condition in data]
        data_json = {'title': 'Popular Cities Forecast', 'date': today, 'cities': cities}
        with open(file='output.json', mode='w') as file:
            json.dump(data_json, file, ensure_ascii=False)

if __name__ == '__main__':
    to_continue = True
    while to_continue:
        data = get_forcast_data()
        get_weather_txt(data)
        get_weather_json(data)
        if input('Would you like to try again?\n').upper() == 'NO':
            to_continue = False





f=get_forcast_data()
print(list(f))
```
